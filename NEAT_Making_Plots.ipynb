```python
#ASSIGNMENT 2

import matplotlib.pyplot as plt
import pickle
import seaborn as sns
import sys, os
import numpy as np
import glob
import pandas as pd

#ACCESS CORRECT FOLDER
ea = 'neat' #ENTER WHICH EA YOU ARE RUNNING ('gaus' or 'uni')
enemy = 26 #ENTER ENEMY

folder = 'EA' + ea + '/engroup%i'%enemy
for n,i in enumerate(glob.glob(folder+'/*[!_gains].csv')):
    test = pd.read_csv(i,names=['best','avg','std'])
    gen = np.linspace(0,len(test['avg'])-1,len(test['avg']))
    #plt.plot(gen,test['best'],label='best')
    plt.plot(gen,test['avg'])#,label='exp%i'%(n+1)))
    plt.fill_between(gen,test['avg'] - test['std'], test['avg'] + test['std'],color='gray', alpha=0.2)

plt.legend()
plt.xlabel('generation')
plt.ylabel('fitness')
```
